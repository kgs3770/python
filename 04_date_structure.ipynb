## list copy

```python
origin_list = [1, 2, 3]
copy_list = origin_list
```

```python
print(origin_list)
print(copy_list)
```

```python
copy_list[0] = 100
print(origin_list)
print(copy_list)
```

```python
origin_list = [1, 2, 3]
copy_list = origin_list[:]

copy_list[0] = 100

print(origin_list)
print(copy_list)
```

```python
o_l = [1, 2, [10, 11]]
c_l = o_l[:]

c_l[0] = 100
c_l[2][0] = -10

print(o_l)
print(c_l)
```

```python
import copy #추가하다

o_l = [1, 2, [10, 11]]
c_l = copy.deepcopy(o_l)

c_l[0] = 100
c_l[2][1] = -11

print(o_l)
print(c_l)

```

## list cmprehension

```python
numbers = list(range(1, 11))
print(numbers)
```

```python
# 모든 데이터를 세제곱
result = []

for number in numbers:
    temp = number ** 3
    result.append(temp)

print(result)
```

```python
result = [number ** 3 for number in numbers]
# result = [number1 ** 3, number2 ** 3, number3 ** 3]
print(result)
```

```python
# 짝수만 고르기
numbers = list(range(1, 31))
result = []

for number in numbers:
    if number % 2 == 0:
        result.append(number)

print(result)
```

```python
numbers = list(range(1, 31))
result = [number for number in numbers if number % 2 == 0]
print(result)
```

```python
# 연습! 모음제거
words = 'my name is hong'

# => my nm s hng
```

```python
# 1. for / if
#vowels = ['a', 'e', 'i', 'o', 'u']
vowels = 'aeiou'
result = []
for char in words:
    # 모음인경우 => 모음이 아닌경우
    if char not in vowels:
         result.append(char)
    # 자음인경우
    # else:
        # result.append(char)

print(result)
print(''.join(result))
```

```python
# 2. list comprehension
vowels = 'aeiou'
result = [char for char in words if char not in vowels]
print(result)
print(''.join(result))
```

```python
# 3. replace 메소드
vowels = 'aeiou'

for char in vowels:
    words = words.replace(char, '')

print(words)
```
