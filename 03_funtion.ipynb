## 함수의 return

- 함수가 return을 만나면 해당값을 반환하고 함수를 종료
- 만약 return이 없는 경우 None을 자동으로 반환
- return은 오직 하나의 객체만 반환한다.

````python
result = my_max(1, 2)
print(result)
#result * 2
````

````python
def my_max2(num1, num2):
    if num1 > num2:
        return num1
    else:
        return num2
````

````python
result = my_max2(1, 2)
print(result)
````

- 기본적으로 함수는 위치를 기준으로 인수를 판단합니다.

````python
def my_max(*numbers):
    # numbers = (1, 2, 3, 4, 5, 6, 7)
    #max_num = 0 #0은 초기값
    max_num = numbers[0] #입장하자마자 첫번째 데이터를 손에 쥠
    for number in numbers:
        if max_num < number:
            max_num = number

    return max_num
````

````python
my_max(1, 2, 3, 4, 5, 6, 7)
````

````python
my_max(-1, -2, -3)
````
