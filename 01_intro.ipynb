## 2.3 논리연산자
- and : 양쪽 모두 True 일 때, True를 반환
- or : 양쪽 모두 False 일 때, False를 반환
- not : 상태를 부정

````python
print(True and True)
print(True and False)
print(False and True)
print(False and False)
````

````python
print(True or True)
print(True or False)
print(False or True)
print(False or False)
````

````python
a = True
print(not a)
````

````python
# 단축평가
print(3 and 5)
print(3 and 0)
print(0 and 5)
print(0 and 0)
````

````python
print(3 or 5)
print(3 or 0)
print(0 or 5)
print(0 or 0)
````
